```python
# importing all neccessary libraries for data analysis and visualization
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import shutil
import tempfile
#loading datasets into jupyter notebook through pandas
df = pd.read_csv('netflix_data.csv')

df
```

```python
df.isnull().sum()# codes describing total empty cell values in each observations
```

```python
#filling missing data in the columns to enhance easy dataset analysis
df['director'] = df['director'].fillna('unknown')# filling all the missin data in director column
df['cast'] = df['cast'].fillna('unknown')# filling all missing dataest in cast column
df['country'] = df['country'].fillna('unknown')# missing data in the country columns 
df['date_added'] = df['date_added'].fillna('unknown')# missing null dtaset in the date column
df
```

```python
df.info()# dataset description after filling the missing values 
```

```python
# dropping two cloumns that are not influential to the analysis and therefore cleaned out
df.drop(['show_id','description'], axis = 1)
```

```python
#utilizing seaborn library for visualization of the ratings distributions distinguishing the x and y axis labels 

sns.countplot(x='rating', data=df, order=df['rating'].value_counts().index, palette = "Reds")
plt.title('Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.figure(figsize=(8,4))
plt.show()
```

```python
# bar chart descriptions of the movie/tvshow count 
sns.countplot(data = df , x = 'type');
```

```python
#codes to reflect the 10 top most watched genres
plt.figure(figsize=(10, 6))
top_genres = df['listed_in'].value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette="Reds_r")
plt.title('Top 10 Most watched Genres')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()
```
